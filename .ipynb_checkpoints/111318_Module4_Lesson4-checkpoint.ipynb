```python
import numpy
from scipy import linalg
```

```python
def lhs_operator(M, N, sigma):
    """
    """
    A = numpy.zeros((M * N, M * N))
    for j in range(N):
        for i in range(M):
            I = j * M + i
            south = I - M
            west = I - 1
            east = I + 1
            north = I + M
            
            if i == 0 and j == 0:
                A[I, I] = 1.0 / sigma + 4.0
                A[I, east] = -1.0
                A[I, north] = -1.0
            elif i == M - 1 and j == 0:
                A[I, I] = 1.0 / sigma + 3.0
                A[I, west] = -1.0
                A[I, north] = -1.0
            elif i == 0 and j == N - 1: 
                A[I, I] = 1.0 / sigma + 3.0
                A[I, south] = -1.0
                A[I, east] = -1.0
            elif i == M - 1 and j == N - 1:
                A[I, I] = 1.0 / sigma + 2.0
                A[I, south] = -1.0
                A[I, west] = -1.0
                
            elif i == 0: 
                A[I, I] = 1.0 / sigma + 4.0
                A[I, south] = -1.0
                A[I, east] = -1.0
                A[I, north] = -1.0
            elif i == M - 1:
                A[I, I] = 1.0 / sigma + 3.0
                A[I, south] = -1.0
                A[I, west] = -1.0
                A[I, north] = -1.0
            elif j == 0:
                A[I, I] = 1.0 / sigma + 4.0
                A[I, west] = -1.0
                A[I, east] = -1.0
                A[I, north] = -1.0
            elif j == N - 1:
                A[I, I] = 1.0 / sigma +3.0
                A[I, south] = -1.0
                A[I, west] = -1.0
                A[I, east] = -1.0
                
            else:
                A[I, I] = 1.0 / sigma + 4.0
                A[I, south] = -1.0
                A[I, west] = -1.0
                A[I, east] = -1.0
                A[I, north] = -1.0
    return A
```

```python
def btcs_implicit_2d(T0, nt, dt, dx, alpha, Tb):
    """
    """
    ny, nx = T0.shape
    M, N = nx - 2, ny - 2
    sigma = alpha * dt / dx**2
    A = lhs_operator(M, N, sigma)
    T = T0[1:-1, 1:-1].flatten()
    I, J = int(M / 2), int(N / 2)
    
    for n in range(nt):
        b = rhs_vector(T, M, N, sigma, Tb)
        T = linalg.solve(A, b)
        if T[J * M + I] >= 70.0:
            break
    print('[time step {}] Center at T = {:.2f} at t={:.2f} s' 
          .format(n+1, T[J * M + I], (n + 1) * dt))
    return map_1d_to_2d(T, nx, ny, Tb)
```

```python
def rhs_vector(T, M, N, sigma, Tb):
    """
    """
    b = 1.0 / sigma * T
    for j in range(N):
        for i in range(M):
            I = j * M + i
            if i == 0:
                b[I] += Tb
            if j == 0:
                b[I] += Tb
    return b
```

```python
def map_1d_to_2d(T_1d, nx, ny, Tb):
    """
    """
    T = numpy.zeros((ny, nx))
    T[1:-1, 1:-1] = T_1d.reshape((ny - 2, nx - 2))
    T[:, 0] = Tb
    T[0, :] = Tb
    T[:, -1] = T[:, -2]
    T[-1, :] = T[-2, :]
    return T
```

```python
Lx = 0.01
Ly = 0.01
nx = 21
ny = 21
dx = Lx / (nx - 1)
dy = Ly / (ny - 1)
alpha = 1e-4

x = numpy.linspace(0.0, Lx, num = nx)
y = numpy.linspace(0.0, Ly, num = ny)

Tb = 100.0
T0 = 20.0 * numpy.ones((ny, nx))
T0[:, 0] = Tb
T0[0, :] = Tb
```

```python
sigma = 0.25
dt = sigma * min(dx, dy)**2 / alpha
nt = 300

T = btcs_implicit_2d(T0, nt, dt, dx, alpha, Tb)
```

```python
from matplotlib import pyplot
%matplotlib inline
```
